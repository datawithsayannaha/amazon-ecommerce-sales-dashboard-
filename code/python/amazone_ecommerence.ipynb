```python
sales['order_month'] = sales['order_date'].dt.to_period('M')
```

```python
sales[['customer_id','order_date','order_month']].head()
```

```python
#Tracking Customer Activity Across Order Month and Cohort Month (customer first purchase month) 
sales['cohort_month'] = sales.groupby('customer_id')['order_month'].transform('min')
sales[['customer_id','order_month','cohort_month']].head(20)
```

```python
#cohort index
sales['cohort_index'] = (
    (sales['order_month'].dt.year - sales['cohort_month'].dt.year) * 12
    + (sales['order_month'].dt.month - sales['cohort_month'].dt.month)
    + 1
)

sales[['customer_id', 'cohort_month', 'order_month', 'cohort_index']].head(10)
```

```python
cohort_data = sales.groupby(['cohort_month', 'cohort_index'])['customer_id'].nunique().reset_index()
cohort_data.head(20)
```

```python
cohort_pivot = cohort_data.pivot(index='cohort_month', columns='cohort_index', values='customer_id')
cohort_pivot
```

```python
retention = cohort_pivot.divide(cohort_pivot.iloc[:, 0], axis=0) * 100
retention = retention.round(2)
retention
```

```python
retention_reset = retention.reset_index()

retention_long = retention_reset.melt(
    id_vars='cohort_month',
    var_name='cohort_index',
    value_name='retention_rate'
)

retention_long['cohort_month'] = retention_long['cohort_month'].astype(str)

retention_long = retention_long.sort_values(
    by=['cohort_month', 'cohort_index']).reset_index(drop=True)

retention_long.to_csv("cohort_analysis.csv", index=False)

retention_long.head(20)
```

```python
#monthly summary (monthly revenue,profit,order trend)
monthly_summary = (
    sales
    .groupby(['order_year', 'order_month'], as_index=False)
    .agg(
        total_orders=('order_id', 'nunique'),
        total_customers=('customer_id', 'nunique'),
        total_revenue=('revenue', 'sum'),
        total_profit=('profit', 'sum'),
        total_quantity=('quantity', 'sum')
    )
)

monthly_summary['avg_order_value'] = (
    monthly_summary['total_revenue'] / monthly_summary['total_orders']
)

monthly_summary.head()
```

```python
monthly_summary.to_csv("monthly_summary.csv",index =False)
```

```python
monthly_summary = monthly_summary.sort_values(['order_year', 'order_month']).reset_index(drop=True)

rolling_metrics = monthly_summary.copy()

rolling_metrics['rolling_3m_revenue'] = (
    rolling_metrics['total_revenue']
    .rolling(window=3, min_periods=1)
    .sum()
)

rolling_metrics['rolling_3m_profit'] = (
    rolling_metrics['total_profit']
    .rolling(window=3, min_periods=1)
    .sum()
)

rolling_metrics['rolling_3m_orders'] = (
    rolling_metrics['total_orders']
    .rolling(window=3, min_periods=1)
    .sum()
)

rolling_metrics.head()
```

```python
rolling_metrics.to_csv("rolling_metrics.csv", index=False)
```

```python
import pandas as pd

events = pd.read_csv("events.csv")
events['event_timestamp'] = pd.to_datetime(events['event_timestamp'])

events['event_type'] = events['event_type'].str.lower().str.strip()

views = events.loc[events['event_type'] == 'view', 'user_id'].nunique()
wishlist = events.loc[events['event_type'] == 'wishlist', 'user_id'].nunique()
carts = events.loc[events['event_type'] == 'cart', 'user_id'].nunique()
purchases = events.loc[events['event_type'] == 'purchase', 'user_id'].nunique()

funnel_data = pd.DataFrame({
    'stage': ['view', 'wishlist', 'cart', 'purchase'],
    'users': [views, wishlist, carts, purchases]
})

# adjusted funnel values
funnel_data['users'] = funnel_data['users'].cummin()

funnel_data['conversion_from_previous'] = [
    100,
    (funnel_data.loc[1, 'users'] / funnel_data.loc[0, 'users'] * 100) if funnel_data.loc[0, 'users'] > 0 else 0,
    (funnel_data.loc[2, 'users'] / funnel_data.loc[1, 'users'] * 100) if funnel_data.loc[1, 'users'] > 0 else 0,
    (funnel_data.loc[3, 'users'] / funnel_data.loc[2, 'users'] * 100) if funnel_data.loc[2, 'users'] > 0 else 0
]

funnel_data['conversion_from_view'] = [
    100,
    (funnel_data.loc[1, 'users'] / funnel_data.loc[0, 'users'] * 100) if funnel_data.loc[0, 'users'] > 0 else 0,
    (funnel_data.loc[2, 'users'] / funnel_data.loc[0, 'users'] * 100) if funnel_data.loc[0, 'users'] > 0 else 0,
    (funnel_data.loc[3, 'users'] / funnel_data.loc[0, 'users'] * 100) if funnel_data.loc[0, 'users'] > 0 else 0
]

funnel_data['conversion_from_previous'] = funnel_data['conversion_from_previous'].round(2)
funnel_data['conversion_from_view'] = funnel_data['conversion_from_view'].round(2)

funnel_data.to_csv("funnel_data.csv", index=False)
print(funnel_data)
```
